```python
import requests
import json
apiURL = "http://127.0.0.1:5000/recommend"
food = input("Enter food name (or press Enter to skip): ") or None
vitamins = input("Enter required vitamins (comma-separated, or press Enter to skip): ")
minerals = input("Enter required minerals (comma-separated, or press Enter to skip): ")
vitamin_list = [v.strip() for v in vitamins.split(",")] if vitamins else []
mineral_list = [m.strip() for m in minerals.split(",")] if minerals else []
protein = float(input("Enter minimum protein (default 1.03): ") or 1.03)
carbs = float(input("Enter maximum carbs (default 6.89): ") or 6.89)
fats = float(input("Enter maximum fats (default 1.658): ") or 1.658)
testInput = {
    "food": food,
    "vitamins": vitamin_list,
    "minerals": mineral_list,
    "protein": protein,
    "carbs" : carbs,
    "fats" : fats
}
response = requests.post(apiURL, json=testInput)
print("\nStatus Code:", response.status_code)
print("Response JSON:", json.dumps(response.json(), indent=4))
```
